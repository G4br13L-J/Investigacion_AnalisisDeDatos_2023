```python
#Encontrar MPV del equipo esta funcion lo que realiza es buscar el mejor jugador por puntos de un equipo ingresado
import pandas as pd

def encontrar_MVP_equipo():
    df = pd.read_csv('/content/2023_nba_player_stats.csv')
    data = pd.DataFrame(df)
    equipo = input("Ingrese un equipo para saber quien es su MVP: ")
    jugadores_equipo = data[data['Equipo'] == equipo]

    # Encontrar al jugador con la mayor cantidad de puntos
    jugador_mvp = jugadores_equipo[jugadores_equipo['Puntos_marcados'] == jugadores_equipo['Puntos_marcados'].max()]

    # Devolver al MVP
    return jugador_mvp[['Nombre', 'Puntos_marcados']]

encontrar_MVP_equipo()
```

```python
import pandas as pd

def comparativaEquipos():
    df = pd.read_csv('/content/2023_nba_player_stats.csv')
    data = pd.DataFrame(df)
    print("Comparativa de Equipos, ingrese 2 equipos de la NBA para compararlos")
    equipo1 = input("Ingrese el equipo 1 de la NBA: ")
    equipo2= input("Ingrese el equipo 2 de la NBA:")
    jugadores_equipo1 = data[data['Equipo'] == equipo1]
    jugadores_equipo2 = data[data['Equipo'] == equipo2]
    puntos_equipo1 = jugadores_equipo1['Puntos_marcados'].sum()
    puntos_equipo2 = jugadores_equipo2['Puntos_marcados'].sum()
    asistencias_equipo1 = jugadores_equipo1['Asistencias'].sum()
    asistencias_equipo2 = jugadores_equipo2['Asistencias'].sum()
    rebotes_equipo1 = jugadores_equipo1['Total_de_rebotes'].sum()
    rebotes_equipo2 = jugadores_equipo2['Total_de_rebotes'].sum()

    #IMPRIMIR EQUIPO CON MAYOR CANTIDAD DE PUNTOS
    print('\n')
    print("Cantidad de puntos de ambos equipos:")
    print(equipo1,": ",puntos_equipo1)
    print(equipo2,": ",puntos_equipo2)
    if puntos_equipo1>= puntos_equipo2:
       print("El equipo que tiene mas cantidad de puntos es:", equipo1)
    else:
       print("El equipo que tiene mas cantidad de puntos es:", equipo2)

    #IMPRIMIR EQUIPO CON MAYOR CANTIDAD DE ASISTENCIAS
    print('\n')
    print("Cantidad de asistencias de ambos equipos:")
    print(equipo1,": ",asistencias_equipo1)
    print(equipo2,": ",asistencias_equipo2)
    if asistencias_equipo1>= asistencias_equipo2:
       print("El equipo que tiene mas cantidad de asistencias es:", equipo1)
    else:
       print("El equipo que tiene mas cantidad de asistencias es:", equipo2)


    #IMPRIMIR EQUIPO CON MAYOR CANTIDAD DE REBOTES
    print('\n')
    print("Cantidad de rebotes de ambos equipos:")
    print(equipo1,": ",rebotes_equipo1)
    print(equipo2,": ",rebotes_equipo2)
    if rebotes_equipo1>= rebotes_equipo2:
       print("El equipo que tiene mas cantidad de rebotes es:", equipo1)
    else:
       print("El equipo que tiene mas cantidad de rebotes es:", equipo2)

comparativaEquipos()
```

```python
#Encontrar jugador mas joven y el mas viejo de un equipo

import pandas as pd

def encontrar_Jugador_Joven_Y_Viejo():
    df = pd.read_csv('/content/2023_nba_player_stats.csv')
    data = pd.DataFrame(df)
    equipo = input("Ingrese un equipo para saber su jugador más joven y el más viejo: ")
    jugadores_equipo = data[data['Equipo'] == equipo]

    # Encontrar al jugador más joven y al más viejo
    jugador_joven = jugadores_equipo[jugadores_equipo['Edad'] == jugadores_equipo['Edad'].min()]
    jugador_viejo = jugadores_equipo[jugadores_equipo['Edad'] == jugadores_equipo['Edad'].max()]

    # Devolver al jugador más joven y al más viejo
    return jugador_joven, jugador_viejo

# Llama a la función para encontrar al jugador más joven y al más viejo de un equipo
jugador_joven, jugador_viejo = encontrar_Jugador_Joven_Y_Viejo()

# Muestra los resultados si el equipo existe
if jugador_joven.empty:
    print("El equipo no existe o no tiene jugadores.")
else:
    print(f"El jugador más joven del equipo es:")
    print(jugador_joven[['Nombre', 'Edad']])

    print(f"El jugador más viejo del equipo es:")
    print(jugador_viejo[['Nombre', 'Edad']])
```

```python
#Mostrar la info del nombre del jugador ingresado.
import pandas as pd

def buscar_jugador_por_nombre():
    # Leer los datos desde el archivo CSV
    df = pd.read_csv('/content/2023_nba_player_stats.csv')

    # Solicitar al usuario ingresar el nombre (o parte del nombre) del jugador
    nombre_jugador = input("Ingrese el nombre (o parte del nombre) del jugador que desea buscar: ")

    # Filtrar los datos para encontrar jugadores cuyo nombre coincida total o parcialmente con el nombre proporcionado
    jugadores_encontrados = df[df['Nombre'].str.contains(nombre_jugador, case=False, na=False)]

    if jugadores_encontrados.empty:
        print("No se encontraron jugadores con ese nombre.")
    else:
        print(f"Resultados de la búsqueda para '{nombre_jugador}':")
        print(jugadores_encontrados[['Nombre', 'Equipo', 'Edad', 'Puntos_marcados', 'Minutos_jugados']])  #Muestra la informacion de las estas variables de los jugadores

# Llamar a la función para buscar jugadores por nombre  #Ejemplo de uso Peyton Watson, Lebron James
buscar_jugador_por_nombre()
```

```python
#Esta función se encarga de encontrar al jugador que más aporta en el trabajo defensivo en toda la NBA, tomando los robos y bloqueos exitosos como referencia.
import pandas as pd

def mejores_jugadores_defensivos():
        df = pd.read_csv('/content/2023_nba_player_stats.csv')
        data = pd.DataFrame(df)

        #Se calcula una métrica de defensa que consiste en la suma de los bloqueos y robos exitosos de cada jugador para luego encontrar al mejor defensor basado en esa métrica.
        data['Métrica_Defensa'] = data['Robos_exitosos'] + data['Bloqueos_exitosos']
        mejor_defensor = data[data['Métrica_Defensa'] == data['Métrica_Defensa'].max()]

        # Devuelve al mejor defensor de toda la NBA
        return mejor_defensor[['Nombre','Posicion','Equipo','Bloqueos_exitosos','Robos_exitosos','Métrica_Defensa']]

# Llama a la función para imprimir mejor defensor de toda la NBA
mejores_jugadores_defensivos()
```

```python
import pandas as pd

# Esta funcion tendrá como objetivo encontrar al jugador que menos puntos marcó durante la temporada
def jugador_con_mas_puntos():
  df = pd.read_csv('/content/2023_nba_player_stats.csv', encoding='latin-1')
  data = pd.DataFrame(df)

  # Aqui se establece cual fue la cantidad de puntos maximos
  puntos = data['Puntos_marcados'].max()

  #Esta instruccion asocia asocia los puntos maximos establecidos con el nombre del jugador
  jugador = data[data['Puntos_marcados']==puntos]['Nombre'].iloc[0]
  # El .iloc[0] se encargara de omitir el numero de posicion del jugador para que solo se incluya su nombre

  # Imprime el nombre del jugador y la cantidad de puntos que marcó
  print("El jugador con mas puntos es: ", jugador)
  print("Con un total de: ", puntos, " puntos marcados")

jugador_con_mas_puntos()
```

```python
import pandas as pd

def jugador_con_menos_minutos_jugados():
  df = pd.read_csv('/content/2023_nba_player_stats.csv', encoding='latin-1')
  data = pd.DataFrame(df)

  # Se establece la cantidad minima de minutos que se encuentran
  minutos = data['Minutos_jugados'].min()

  # Se asocian los minutos respectivamente con el nombre del jugador
  jugador = data[data['Minutos_jugados']==minutos]['Nombre'].iloc[0]

  # Se imprime el nombre del jugador que jugó menos tiempo junto con la cantidad de minutos
  print("El jugador que menos minutos ha jugados es: ", jugador)
  print("Con un total de: ", minutos, " minutos jugados")

jugador_con_menos_minutos_jugados()
```

```python
#Para mostar los triples de los jugadores con un grafico
import pandas as pd
import plotly.express as px

df = pd.read_csv('/content/2023_nba_player_stats.csv')

top_players = df[['Nombre', 'Triples_marcados']].sort_values(by='Triples_marcados', ascending=False).head(10)

# Crear un gráfico de barras con Plotly
fig = px.bar(top_players, x='Nombre', y='Triples_marcados', title='Top 10 Jugadores con Más Triples Marcados')
fig.update_xaxes(title_text='Nombre del Jugador')
fig.update_yaxes(title_text='Triples Marcados')

# Mostrar el gráfico
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el DataFrame desde el archivo CSV
df = pd.read_csv('/content/2023_nba_player_stats.csv')

# Calcular el promedio de edad de cada equipo
avg_age_by_team = df.groupby('Equipo')['Edad'].mean().reset_index()

# Crear un gráfico de barras
plt.figure(figsize=(12, 6))  #para el tamaño del grafico
plt.bar(avg_age_by_team['Equipo'], avg_age_by_team['Edad'])#crea el grafico
plt.title('Promedio de Edad de los jugadores de cada equipo')
plt.xlabel('Equipo')
plt.ylabel('Promedio de Edad')
plt.xticks(rotation=45)  # Inclinacion de los nombres para mayor legibilidad

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sms

df = pd.read_csv('/content/2023_nba_player_stats.csv')

contingency_table = pd.crosstab(df['Edad'], df['Posicion'])

plt.figure(figsize = (10, 7))
sms.heatmap(contingency_table, annot = True, cmap = "coolwarm")

plt.title('Mapa de calor de Edad vs Posición')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('/content/2023_nba_player_stats.csv')

# Calcular la frecuencia de cada categoría en 'Posicion'
position_counts = df['Posicion'].value_counts()

# Crear un gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de porcentajes de la posición de Jugadores de la NBA en 2023')
plt.show()
```
